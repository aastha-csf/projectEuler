#### Multiples of 3 or 5 - Problem 1
##### Question
If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.
##### Logic (Brute Force)
1. Iterate through all numbers from `1` to `n` = `1000`
2. For each number `i`, if `i` is divisible by 3 or 5, add it to `sumOfMultiples`
##### Optimized Approach
`Sum of an arithmetic series = `$\frac{(k(k+1))}{2}$
1. Instead of iterating through a loop use the formula to find the sum
2. Above is only the sum of `(1 + 2 + 3 + 4 + ...)` but we want to `Find the sum of all the multiples of 3 or 5 below 1000.`

*To find all the multiples of 3 until n*

$$
\begin{gathered}
3 + 6 + 9 + 12 + 15 + 18 + 21 + \ldots + 999 \\
(3\cdot1) + (3\cdot2) + (3\cdot3) + (3\cdot4) + \ldots + (3\cdot333)
\end{gathered}
$$

<p style="text-align: center;">Factor out the 3</p>

$$
    3\cdot(1+2+3+4+5+6+7+\ldots+333)
$$
    
<p style="text-align: center;">Sum of the multiples of 3 = 3 * ("Sum of an arithmetic series")</p>

$$3(\frac{k(k+1)}{2})$$

*Follow the same to find all the multiples of 5*

3. Observe that the multiples of 3 and 5 overlap so to remove one of the pair: `-sumOfTheMultiplesOf(15)`

*Multiples of 3:* `3 + 6 + 9 + 12 +` __15__ `+ 18 + 21 + 24 + 27 +` __30__ `+ 33`  
*Multiples of 5:* `5 + 10 +` __15__ `+ 20 + 25 +` __30__ `+ 35`

4. Sum of all the multiples of 3 or 5 : sum of Multiples of 3 + sum of Multiples of 5 - `sumOfTheMultiplesOf(15)`

<p style="text-align: center;">A = Multiples of 3<br>B = Multiples of 5</p>

$$
\sum_{} A \cup B = \sum_{} A + \sum_{} B - \sum_{} A \cap B
$$

##### Time & Space Complexity
- **Time:** \(O(1)\)
- **Space:** \(O(1)\)
